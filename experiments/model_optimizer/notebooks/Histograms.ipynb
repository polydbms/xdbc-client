```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/grid_configurations_FIXED.csv")
#data = pd.read_csv('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C16_N1000_random_samples.csv')
bins=5

columns_to_plot = ["compression","bufpool_size","buffer_size","send_par","write_par","decomp_par","read_partitions","read_par","deser_par","ser_par","comp_par"]

for column in data.columns:
        if column not in columns_to_plot:
            continue
    
        value_counts = data[column].value_counts()

        if pd.api.types.is_numeric_dtype(data[column]):
                value_counts = value_counts.sort_index()
        
        plt.figure()
        value_counts.plot(kind='bar', edgecolor='k', alpha=0.7)
        plt.title(f'Occurrences of Unique Values in {column}')
        plt.xlabel('Unique Values')
        plt.ylabel('Number of Occurrences')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
```
