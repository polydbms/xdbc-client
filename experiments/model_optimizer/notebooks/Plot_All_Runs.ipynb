```python
import numpy as np
from scipy.stats import bootstrap
import pandas as pd
import glob
from collections import defaultdict
import typing
import matplotlib.pyplot as plt


file_list = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/results_1310k/*/cost*/*.csv')

env = ""
algo = ""

for file in file_list:
    try:

        df = pd.read_csv(file)
        df = df[df['time'] > 0]
        trial_id = df['trial_id'][0]
    
        #if trial_id != 85:
        #    continue
        df = df[df['transfer_id'] != -1]

        
        env = file.split("/")[-1].split("\\")[-3]
        algo = file.split("\\")[-2]
    
        #if "clustering" not in algo:
        #        continue

        #if "changing" not in algo:
        #        continue
    
        
    
        plt.figure(figsize=(10, 6))
        plt.plot(df['time'],label="time")
        plt.plot(df['predicted_time'],label="predicted_time")
        plt.xlabel('Itertions')
        plt.ylabel('Runtime seconds')

        if env == "S2_C2_N50":
            plt.axhline(y=76, color='red', linestyle='--', label='Best Time S2_C2_N50')
            plt.axhline(y=122, color='black', linestyle='--', label='Average Time S2_C2_N50')

        if env == "S2_C2_N150":
            plt.axhline(y=50, color='red', linestyle='--', label='Best Time S2_C2_N150')
            plt.axhline(y=78, color='black', linestyle='--', label='Average Time S2_C2_N150')

        if env == "S2_C2_N1000":
            plt.axhline(y=43, color='red', linestyle='--', label='Best Time S2_C2_N1000')
            plt.axhline(y=72, color='black', linestyle='--', label='Average Time S2_C2_N1000')

        
        if env == "S8_C2_N50":
            plt.axhline(y=69, color='red', linestyle='--', label='Best Time S8_C2_N50')
            plt.axhline(y=118, color='black', linestyle='--', label='Average Time S8_C2_N50')

        if env == "S8_C2_N150":
            plt.axhline(y=32, color='red', linestyle='--', label='Best Time S8_C2_N150')
            plt.axhline(y=53, color='black', linestyle='--', label='Average Time S8_C2_N150')

        if env == "S8_C2_N1000":
            plt.axhline(y=22, color='red', linestyle='--', label='Best Time S8_C2_N1000')
            plt.axhline(y=42, color='black', linestyle='--', label='Average Time S8_C2_N1000')

        
        if env == "S16_C16_N50":
            plt.axhline(y=68, color='red', linestyle='--', label='Best Time S16_C16_N50')
            plt.axhline(y=119, color='black', linestyle='--', label='Average Time S16_C16_N50')

        if env == "S16_C16_N150":
            plt.axhline(y=28, color='red', linestyle='--', label='Best Time S16_C16_N150')
            plt.axhline(y=53, color='black', linestyle='--', label='Average Time S16_C16_N150')

        if env == "S16_C16_N1000":
            plt.axhline(y=20, color='red', linestyle='--', label='Best Time S16_C16_N1000')
            plt.axhline(y=41, color='black', linestyle='--', label='Average Time S16_C16_N1000')

        plt.ylim(bottom=0)
        #plt.ylim(0, 100)
        plt.title(f"Times chronologicaly for Environment {env} and Algorithm {algo} with Trial ID {trial_id}")
        plt.grid()
        plt.legend()
        plt.show()
    except:
        pass
```
