```python
import pandas as pd
import glob
import yaml


#all_files = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/results_6250k/*/*/*.csv')

all_files = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/results_1310k/*/*/*.csv')
#all_files = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/results_1310k_shorter_runs/*/*/*.csv')
#all_files = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/results_175k/*/*/*.csv')




all_envs = []
all_algos = []
total_count = 0
dict_overview = dict();


dict_count_per_env = dict();

for file in all_files:
    df = pd.read_csv(file)

    env = file.split("\\")[-3]
    algo = file.split("\\")[-2]
    count = len(df[df['time'] > 0])

    total_count = total_count + count

    #print("Env : "+str(env)+"   Algo : "+str(algo)+"   Count : "+str(count))
    
    if env not in dict_overview.keys() :
        dict_overview[env] = dict()
    if algo not in dict_overview[env].keys(): 
        dict_overview[env][algo] = ""

    if env not in dict_count_per_env.keys():  
        dict_count_per_env[env] = 0
    
    dict_overview[env][algo] = dict_overview[env][algo] + "" + str(count)+"x;"
    dict_count_per_env[env] = dict_count_per_env[env] + count

print(yaml.dump(dict_overview))
print(yaml.dump(dict_count_per_env))
print("Total Transfer Count : "+str(total_count))
```
