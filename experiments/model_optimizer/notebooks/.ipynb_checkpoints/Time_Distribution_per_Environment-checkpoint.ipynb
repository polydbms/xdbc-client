```python

'''

Plot The Time Distriution for each file list using only random samples

'''


import glob
import pandas as pd
import seaborn as sns
import statistics
import numpy as np
import pandas as pd
import seaborn as sns
import glob, os
import matplotlib.pyplot as plt

file_list_env_1 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C16_N1000_random_sample*.csv')
file_list_env_2 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C8_N100_random_sample*.csv')
file_list_env_3 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S8_C1_N50_random_sample*.csv')
file_list_env_4 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C16_N10000_random_sample*.csv')
file_list_env_5 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C4_N10_random_sample*.csv')
file_list_env_6 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S8_C8_N500_random_sample*.csv')
file_list_env_7 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S2_C2_N20_random_sample*.csv')
file_list_env_8 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C1_N1000_random_sample*.csv')
file_list_env_9 = glob.glob('C:/Users/bened/Desktop/Uni/repos/xdbc-client/experiments/model_optimizer/random_samples_1310k/S16_C1_N10_random_sample*.csv')

file_lists_1310k = [file_list_env_1,file_list_env_2,file_list_env_3,file_list_env_4,file_list_env_5,file_list_env_6,file_list_env_7,file_list_env_8,file_list_env_9]


for file_list in file_lists_1310k:

    env = ""
    algo = ""

    all_data = pd.DataFrame()
    list_single_dfs_1310k = []

    for file in file_list:

        env = file.split("/")[-1].split("\\")[1].split("_random")[0]
        algo = file.split("\\")[-2]

        df = pd.read_csv(file)
        df = df[df['time'] > 2]

        list_single_dfs_1310k.append(df)
        
    all_data = pd.concat(list_single_dfs_1310k, axis=0, ignore_index=True)

    all_data = all_data.sort_values(by=['time'],ascending=False).reset_index()
    
    print(f"Max transfer time for env {env} :     {all_data['time'].max()}")
    print(f"Average transfer time for env {env} : {statistics.mean(all_data['time'])}")
    print(f"Min transfer time for env {env} :     {all_data['time'].min()}")
    
    plt.figure(figsize=(8, 5))
    
    plt.plot(all_data['time'])
    plt.grid()
    plt.xlabel('Configurations')
    plt.ylabel('Time seconds')
    plt.ylim(bottom=0)
    plt.title(f"Time Distribution for Environment {env}")# and Algorithm {algo}")
    plt.show()
    #plt.savefig(f'time_distribution_{env}.png', dpi=300, bbox_inches='tight')

    


```
